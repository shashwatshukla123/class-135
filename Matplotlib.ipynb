```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([0,100,150,250])
# 'D:b':-- marker[o],line[:],color[b]
plt.plot(ypoints,'D:k')
plt.show()
```

Line  Syntax	Description

'-'	  Solid line [linestyle='solid']	
\
':'	  Dotted line	[linestyle='dotted']
\
'--'  Dashed line	[linestyle='dashed']
\
'-.'  Dashed/dotted line [linestyle='dashdot']
\
'or'  none [linestyle='None']

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0,3,5, 6])
ypoints = np.array([0,100,150,250])
# 'D:b':-- marker[o],line[:],color[b]
# ms for markersize.
# mec for markeredgecolor.
# mfc for markerfacecolor.
# linestyle for style of the line.
#linewidth for width of the line.
# color for the color of the line.
plt.plot(ypoints,'D:k',ms=20, mec='r',mfc='g',linestyle='solid',linewidth = '10.5',color='k')
plt.plot(xpoints,'*',ms=20, mec='r',mfc='g',ls='dotted',lw = '10.5', c = '#4CAF50')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

#fontdict:- to set font properties for the title and labels.
# loc= for the location title [Legal values are: 'left', 'right', and 'center[default]'].
plt.title("Sports Watch Data",fontdict = font1,loc = 'left')
plt.xlabel("Average Pulse",fontdict = font2)
plt.ylabel("Calorie Burnage",fontdict = font2)

plt.plot(x1, y1, x2, y2)
plt.show()
```
